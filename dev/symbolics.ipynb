# Symbolic Form of DH/MDH and Jacobian

```python
import sympy as sp
from sympy import init_printing
from sympy import init_session

init_session();
init_printing();
```

```python
# Define symbolic joint angles
theta = sp.symbols('theta1:7')  # theta1..theta6

# Define symbolic DH parameters
a = sp.symbols('a1:7')          # a1..a6
d = sp.symbols('d1:7')          # d1..d6
alpha = sp.symbols('alpha1:7')  # alpha1..alpha6

def create_dh_matrix(theta, d, a, alpha, modified_DH=True):
    ct = 1 #sp.cos(theta)
    st = 1#sp.sin(theta)
    ca = sp.cos(alpha)
    sa = sp.sin(alpha)
    
    if modified_DH:
        return sp.Matrix([
            [ct, -st, 0, a],
            [st*ca, ct*ca, -sa, -d*sa],
            [st*sa, ct*sa, ca, d*ca],
            [0, 0, 0, 1]
        ])
    else:
        return sp.Matrix([
            [ct, -st*ca, st*sa, a*ct],
            [st, ct*ca, -ct*sa, a*st],
            [0, sa, ca, d],
            [0, 0, 0, 1]
        ])
```

```python
frames = [sp.eye(4)]
for i in range(len(theta)):
    T = create_dh_matrix(theta[i], a[i], d[i], alpha[i])
    frames.append(frames[-1] * T)

# End-effector position
end_effector_pos = frames[-1][:3, 3]
print("End-effector position:")
print("Coor ee_pos{", ",\r\n".join([sp.ccode(p) for p in end_effector_pos]), "};")
```

```python
dh_params = [
    (167.719, 0, 0.0),
    (0, 0, sp.pi/2),
    (0, 250.201, 0 ),
    (231.03, 0, sp.pi/2),
    (0, 0, -sp.pi/2),
    (156.927, 0, sp.pi/2)
]

frames = [sp.eye(4)]
for i in range(len(theta)):
    T = create_dh_matrix(theta[i], *dh_params[i])
    frames.append(frames[-1] * T)


end_effector_pos = frames[-1][:3, 3]
print("End-effector position:")
print("Coor ee_pos{", ",\r\n".join([sp.ccode(p) for p in end_effector_pos]), "};")
```

```python

# zyz
thetaz = sp.symbols('thetaz')
thetay = sp.symbols('thetay')
thetax = sp.symbols('thetax')

phi = sp.symbols('phi')
psi = sp.symbols('psi')
theta = sp.symbols('theta')

# rx = sp.Matrix([[1, 0, 0],
#                 [0, sp.cos(thetax),-sp.sin(thetax)],
#                 [0, sp.sin(thetax), sp.cos(thetax)]])

rytheta = sp.Matrix([[sp.cos(theta), 0, sp.sin(theta)],
                     [0, 1, 0],
                     [-sp.sin(theta), 0, sp.cos(theta)]])

rzpsi = sp.Matrix([[sp.cos(psi), -sp.sin(psi), 0],
                   [sp.sin(psi), sp.cos(psi), 0],
                   [0, 0, 1]])

rzphi = sp.Matrix([[sp.cos(phi), -sp.sin(phi), 0],
                    [sp.sin(phi), sp.cos(phi), 0],
                    [0, 0, 1]])

R = rzphi * rytheta * rzpsi
print("Matrix3d R{", ",\r\n".join([sp.ccode(sp.simplify(p)) for p in R]), "};")
R
```
